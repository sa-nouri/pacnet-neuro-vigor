```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import seaborn as sns; sns.set_theme()
```

```python
path_pacs_dataset = "/path-to-pacs-dir/"
df = pd.read_parquet(path_pacs_dataset + "data_merged.parquet")
print(df.info)
print(df)
```

```python
df_sham = df[df["stimulus"] == "sham"]
df_sham_hc = df_sham[df_sham["medication"] == "hcoffmed"]
print("Get the unique values of medication:", df_sham["medication"].unique())
```

```python
df_sham_hc["beta_gamma"] = df_sham_hc["pac_values"].apply(lambda rec: rec.reshape(60, 30)[32:45, 16:].ravel())
df_sham_hc["alpha_gamma"] = df_sham_hc["pac_values"].apply(lambda rec: rec.reshape(60, 30)[32:45, 8:16].ravel())
df_sham_hc["theta_gamma"] = df_sham_hc["pac_values"].apply(lambda rec: rec.reshape(60, 30)[32:45, 4:8].ravel())
df_sham_hc["delta_gamma"] = df_sham_hc["pac_values"].apply(lambda rec: rec.reshape(60, 30)[32:45, :4].ravel())

df_sham_hc["alpha_beta"] = df_sham_hc["pac_values"].apply(lambda rec: rec.reshape(60, 30)[16:32, 8:16].ravel())
df_sham_hc["theta_beta"] = df_sham_hc["pac_values"].apply(lambda rec: rec.reshape(60, 30)[16:32, 4:8].ravel())
df_sham_hc["delta_beta"] = df_sham_hc["pac_values"].apply(lambda rec: rec.reshape(60, 30)[16:32, :4].ravel())

df_sham_hc["theta_alpha"] = df_sham_hc["pac_values"].apply(lambda rec: rec.reshape(60, 30)[8:16, 4:8].ravel())
df_sham_hc["delta_alpha"] = df_sham_hc["pac_values"].apply(lambda rec: rec.reshape(60, 30)[8:16, :4].ravel())

df_sham_hc["delta_theta"] = df_sham_hc["pac_values"].apply(lambda rec: rec.reshape(60, 30)[4:8, :4].ravel())
```

```python
beta_gamma = np.concatenate(df_sham_hc["beta_gamma"].to_numpy(), axis=0)
alpha_gamma = np.concatenate(df_sham_hc["alpha_gamma"].to_numpy(), axis=0)
theta_gamma = np.concatenate(df_sham_hc["theta_gamma"].to_numpy(), axis=0)
delta_gamma = np.concatenate(df_sham_hc["delta_gamma"].to_numpy(), axis=0)

alpha_beta = np.concatenate(df_sham_hc["alpha_beta"].to_numpy(), axis=0)
theta_beta = np.concatenate(df_sham_hc["theta_beta"].to_numpy(), axis=0)
delta_beta = np.concatenate(df_sham_hc["delta_beta"].to_numpy(), axis=0)

theta_alpha = np.concatenate(df_sham_hc["theta_alpha"].to_numpy(), axis=0)
delta_alpha = np.concatenate(df_sham_hc["delta_alpha"].to_numpy(), axis=0)

delta_theta = np.concatenate(df_sham_hc["delta_theta"].to_numpy(), axis=0)
```

```python
## Beta Gamma

bins = 200
sns.set_style("white", 
            {'font.family':'serif', 
            'font.serif':'Times New Roman'})

fig = plt.figure(figsize=(10, 6), tight_layout=True)

plt.hist(beta_gamma, bins=bins, color='b')

plt.ylabel("Frequency", fontweight='bold', fontsize=12)
plt.xlabel('PAC values', fontweight='bold', fontsize=12)
plt.title("Beta Gamma PAC distribution", fontsize=15, fontweight='bold')
```
